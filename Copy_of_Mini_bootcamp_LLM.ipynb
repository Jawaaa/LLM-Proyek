```python
from groq import Groq
import getpass

GROQ_API_KEY=getpass.getpass("Groq API Key: ")

client = Groq(api_key=GROQ_API_KEY)
```

```python
system_message = f"""
Kamu asisten intelijen yang super helpful.
Jawabalah dengan singkat, padat, dan langsung ke poin dengan gaya santai.
"""
```

```python
def get_response(query: str, context: str, max_length: int = 3000) -> str:
    try:
        # Potong context jika terlalu panjang
        if len(context) > max_length:
            context = context[:max_length] + "..."

        # Panggil API untuk menghasilkan respons
        completion = client.chat.completions.create(
            model="llama-3.2-11b-vision-preview",  # Model yang digunakan
            messages=[
                {"role": "system", "content": system_message},
                {"role": "user", "content": context},
                {"role": "user", "content": query},
            ],
            temperature=0,  # Respons lebih deterministik
            max_tokens=500,  # Batas maksimal panjang respons
            top_p=1,  # Sampling berbasis probabilitas
            stream=False,
            stop=None,
        )

        # Ambil respons dari API
        return completion.choices[0].message.content

    except Exception as e:
        # Tangani error dan tampilkan pesan yang lebih ramah
        return f"❌ Error: {str(e)}"
```

```python
answer = get_response(
    query="jelaskan apa itu dicoding",
    context=context,
    )

print(answer)
```

```python
print(
    get_response(
    query="who is tutor in dicoding",
    context=context,
    )
)
```

```python
print(
    get_response(
    query="what program dunia coding have",
    context=context,
    )
)
```

```python
import gradio as gr

def chat_interface(query, context):
    try:
        response = get_response(query, context)
        return response
    except Exception as e:
        return f"Error: {str(e)}"

# Gradio Interface
iface = gr.Interface(
    fn=chat_interface,
    inputs=[
        gr.Textbox(label="Your Query")
    ],
    outputs=gr.Textbox(label="Assistant Response"),
    title="LLM Chat Interface",
    description="Chat with the LLM"
)

# For Colab, use share=True
iface.launch(pwa=True, share=True)
```
